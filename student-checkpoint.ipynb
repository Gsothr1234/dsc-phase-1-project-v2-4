```python
# Your code here - remember to use markdown cells for comments as well!
#import sqlite3 module and pandas
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
#create connection and cursor to SQlite database for imdb
conn = sqlite3.connect('zippedData/im.db/im.db')
cur = conn.cursor()
```

```python
#For the next step we preview the contents for the movie_basics table
table_names = '''SELECT name as 'Table Name' FROM sqlite_master;'''
pd.read_sql(table_names,conn)
```

```python
# For the next step we view the data in table movie_basics and save it as a Dataframe named df1
query1 = '''SELECT * FROM movie_basics'''
df1 = pd.read_sql(query1,conn)
df1
```

Here we see we have the columns movie_id,primary_title,original_title,start_year,runtime_minutes and genres of note is thta primary_title is not always original_title and not all values for runtime and genre are filled

```python
# For the next step we view the data in table movie_ratings
query2 = '''SELECT * FROM movie_ratings'''
df2 = pd.read_sql(query2,conn)
df2
```

Next step is to merge the two dataframes as we can see they share the same column movie_id also to note is that the dataframes are not the same shape as about half the movies in the imdb database appear to not have ratings

```python
# First we can see for both data_sets movie_id is not the index

imdb = pd.merge(df1,df2, on ='movie_id')
imdb.set_index('movie_id', inplace=True)
imdb
```

Next we view the box office mojo database

```python
df3 = pd.read_csv('zippedData/bom.movie_gross.csv.gz')
df3
```

```python
#Next we merge the two tables 
combdf = imdb.merge(df3, left_on='primary_title',right_on='title')
combdf.drop('title',axis=1,inplace=True)
combdf
```

We can see we have duplicates

```python
#Check datatypes of columns in dataframe
combdf.dtypes
```

```python
#Next is to create a new column for total revenue firstidentify structure of column foreign_gross
q=combdf['foreign_gross'].unique()
print(q)
```

```python
# convert foreign_gross to int
combdf['foreign_gross'] = combdf['foreign_gross'].astype(str)
combdf['foreign_gross'] = combdf['foreign_gross'].str.replace(',','')
combdf['foreign_gross'] = combdf['foreign_gross'].replace('nan','0')
combdf['foreign_gross']= combdf['foreign_gross'].astype(float)
#Create new total_gross column
combdf['total_gross']= combdf['foreign_gross'] + combdf['domestic_gross']
combdf
```

```python
#Next check for duplicates
duplicate = combdf[combdf.duplicated()]
duplicate
```

Next we will attempt to see which genres achieve the highest revenue and thus more likely to be profitable if Microsoft attempt

```python
#Check number of different unique genres
q=combdf['genres'].nunique()
print(q)
combdf.head()
```

```python
#First consider only the first word in genre as the primary identifier
combdf['first_genres'] = combdf['genres'].str.split(',').str[0]

#Group by genre then sum to get total revenue per genre
genres_totals = combdf.groupby('first_genres')['total_gross'].sum().reset_index()

#Create Barplot
plt.figure(figsize =(10,6))
plt.bar(genres_totals['first_genres'], genres_totals['total_gross'])
plt.xlabel('Genres')
plt.ylabel('Total Gross Revenue')
plt.title('Total Gross Revenue by Genre')
plt.xticks(rotation=45)
plt.show()

```

This barchart gives us an idea of the types of movies that are most likely to succeed with action, adventure comedy and drama being the biggest earners

We can further explore this data by viewing a barcount of movies by genre

```python
#calculate movie count by genre
genres_count = combdf['first_genres'].value_counts().reset_index()
genres_count.columns = ['first_genres', 'movie_count']
#Create Barplot
plt.figure(figsize =(10,6))
plt.bar(genres_count['first_genres'], genres_count['movie_count'])
plt.xlabel('Genres')
plt.ylabel('No of Movies')
plt.title('No of Movies by Genre')
plt.xticks(rotation=45)
plt.show()
```

here we can see that surprisingly Action is not the highest despite raising the largest revenue by a significant margin meaning some films are very high earners leading to the graph we saw earlier the most made films are drama action comedy and biography

Next we can check the overall growth state of the industry by tracking revenue per year

```python
#Group by year then sum to get total revenue per year
revenue_per_year = combdf.groupby('year')['total_gross'].sum().reset_index()

#Create linegrap
plt.figure(figsize =(10,6))
plt.plot(revenue_per_year['year'], revenue_per_year['total_gross'])
plt.xlabel('Year')
plt.ylabel('Total Gross Revenue')
plt.title('Total Gross Revenue by Year')
plt.show()
```

It seems the growth is not as stable as expected however despite dips there is a general upward trend.
